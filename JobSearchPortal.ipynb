```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
# JobTitle=[]
# CompanyName=[]
# KeySkills=[]
# Experience=[]
# JobLocation=[]
# JobSalary=[]
# JobDescription=[]

#html_text= requests.get('https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=work%20from%20home&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=work%20from%20home&pDate=I&sequence=1&startPage=0').text
url2="https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=Non%20IT&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=non%20it&pDate=I&sequence=2&startPage="
url= "https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=work%20from%20home&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=work%20from%20home&pDate=I&sequence=1&startPage="
url3="https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=it&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=it&pDate=I&sequence=2&startPage="
def Scrape_Url(url,max_pages):
    
    current_page = 0
    count = 0
    while current_page<= max_pages:
        current_url= url+str(current_page)
        #print(current_url)
        html_text=requests.get(current_url).text
        Scrapping(html_text)
        count+=1
        print(count)
        current_page= current_page+1
    
def Scrapping(html_text): 
    soup= BeautifulSoup(html_text, 'lxml')
    jobs= soup.find_all('li', class_="clearfix job-bx wht-shd-bx")
    for job in jobs:
        #---------------Company Name---------------------------------------------------
        company= job.find('h3', class_= 'joblist-comp-name').text.replace(' ', '').strip()
        b= company.count("(MoreJobs)")
        if b>=1:
            cmp_name= company.replace('(MoreJobs)', '').strip()
#             CompanyName.append(cmp_name)
        else:
#             CompanyName.append(company)
            cmp_name = company


        #----------------Skills-------------------------------------------------------------
        skills= job.find('span', class_= 'srp-skills').text.replace(' ', '').strip()
#         KeySkills.append(skills)

        #----------------Job Title-------------------------------------------------------------
        title= job.find('a').text.replace(' ', ' ').strip()
#         JobTitle.append(title)


         #---------------Experience-------------------------------------------------------------
        ex= job.find('ul', class_= 'top-jd-dtl clearfix') 
        exp= ex.find('li' ).text.replace('card_travel', '').strip()
#         Experience.append(exp)


        #---------------Job Location-------------------------------------------------------------
        lo= job.find('ul', class_= 'top-jd-dtl clearfix') 
        #print(lo)
        loc= ex.find('span' ).text.strip()
#         JobLocation.append(loc)

        #---------------Salary-------------------------------------------------------------
        sa= job.find('ul', class_= 'top-jd-dtl clearfix') 

        sal= ex.find_all('i', class_= 'material-icons rupee')

        if not sal:
            ok='None'
            sal = ok
#             JobSalary.append(ok)
        else:
            repo =job.find('ul',class_="top-jd-dtl clearfix").find_all('li')[1]
            sal = repo.text.strip()
#             JobSalary.append(repo.text.strip())

        #---------------Job Description-------------------------------------------------------------
        d= job.find('ul', class_= 'list-job-dtl clearfix') 
        des=d.find('li').text.replace('Job Description:', '').strip()
#         JobDescription.append(des)
        
        data = {'Job Title':[title],'Company Name':[cmp_name],'Job Location':[loc], 'Job Salery':[sal] , 'Job Description':[des],'Key Skills':[skills],'Experience':[exp]}
        df = pd.DataFrame(data) 
        df.to_csv('Salery1.csv', mode='a',index=False, header=False, encoding='utf-8')
    
#Scrapping(html_text) 
#Max Pages would be approx 9 thousandd
#Scrape_Url(url,8116) 

#Max Pages would be approx 6596 thousand
#Scrape_Url(url2,3400) 

#Max Pages would be approx 6596 thousand
Scrape_Url(url3,3000) 

#print(len(KeySkills))
#print(len(CompanyName))
#print(len(JobTitle))
#print(len(Experience))
#print(len(JobLocation))
#print(len(JobSalary))
#print(len(JobDescription))

```
